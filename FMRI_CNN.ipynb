```python
import pandas as pd
import numpy as np

def reconstruct_symmetric_df(df):
    n = 199  # Since the original upper triangular part corresponds to a 199x199 matrix
    full_matrix_shape = (df.shape[0], n, n)
    full_matrix = np.zeros(full_matrix_shape)

    # Extract upper triangular indices robustly
    indices = []
    for col in df.columns:
        col = col.strip()
        try:
            i, j = int(col.split('throw_')[0]), int(col.split('throw_')[1].split('thcolumn')[0]) - 1
            indices.append((i, j))
        except ValueError:
            raise ValueError(f"Unexpected column format: {col}")

    # Populate upper triangular part of the matrix
    for k, (i, j) in enumerate(indices):
        full_matrix[:, i, j] = df.iloc[:, k].values

    # Mirror to lower triangular part (excluding diagonal, since it's already assigned)
    for i in range(n):
        for j in range(i+1, n):  # Only upper triangle has values
            full_matrix[:, j, i] = full_matrix[:, i, j]

    # Reshape to match required format (1213, 199*199)
    reshaped_df = pd.DataFrame(full_matrix.reshape(df.shape[0], -1))
    return reshaped_df

symmetric_fmri = reconstruct_symmetric_df(df_fmri)
print(symmetric_fmri.head(5))
symmetric_fmri.shape
```

```python
# prompt: open file in TRAIN folder called "TRAINING_SOLUTIONS.xlsx", for ids in participant_ids_list match the ids to TRAINING_SOLUTIONS participate_id column and return the matched Sex_F label list

import pandas as pd

# Assuming the file path is correct and the file exists
solutions_file_path = '/content/drive/MyDrive/WiDS - Team Conscious/data/TRAIN/TRAINING_SOLUTIONS.xlsx'

try:
  solutions_df = pd.read_excel(solutions_file_path)
  print("Solutions file loaded successfully.")
except FileNotFoundError:
  print(f"Error: Solutions file not found at {solutions_file_path}")
except Exception as e:
  print(f"An unexpected error occurred while loading solutions file: {e}")
```

```python
sex_f_labels = []

for participant_id in participant_ids_list:
  matched_row = solutions_df.loc[solutions_df['participant_id'] == participant_id]
  if not matched_row.empty:
    sex_f_label = matched_row['Sex_F'].iloc[0]
    sex_f_labels.append(sex_f_label)
  else:
    print(f"Warning: Participant ID {participant_id} not found in TRAINING_SOLUTIONS.xlsx")
    sex_f_labels.append(None)

assert len(sex_f_labels) == len(participant_ids_list)
```

```python
# https://www.tensorflow.org/tutorials/images/cnn
from tensorflow.keras import datasets, layers, models

model = models.Sequential()
model.add(layers.Conv2D(256, (3, 3), activation='relu', input_shape=(199, 199, 1)))  # Adjusted for grayscale-like input
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(128, (3, 3), activation='relu'))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(64, (3, 3), activation='relu'))
model.add(layers.Flatten())
model.add(layers.Dense(64, activation='relu'))
model.add(layers.Dropout(0.5))
model.add(layers.Dense(2, activation='softmax'))

model.summary()
```

```python
import tensorflow as tf
model.compile(optimizer='adam',
              loss=tf.keras.losses.CategoricalCrossentropy(from_logits=True),
              metrics=['accuracy'])
```
